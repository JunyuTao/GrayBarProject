```python
import pandas as pd
import datetime
```

```python
df_backlog = pd.read_csv('churn_umsl_file2_backlog.csv')
```

```python
df_backlog['new_yearMonth'] = pd.to_datetime(df_backlog['End_of_Month'].astype(str), format='%Y%m%d')
```

```python
df_backlog['new_yearMonth'] = df_backlog['new_yearMonth'].to_numpy().astype('datetime64[M]')
```

```python
df_backlog = df_backlog.drop(columns=['End_of_Month'])
df_backlog
```

```python
df_main = pd.read_csv('combined_csv.csv')
df_main
```

```python
df_main['new_yearMonth'] = pd.to_datetime(df_main['YearMonth'].astype(str), format='%Y%m')
```

```python
df_gb = df_backlog.groupby(['customer_number','new_yearMonth']).sum()
df_gb = df_gb.reset_index()
```

```python
df_gb
```

```python
df_merged = pd.merge(df_main, df_gb, how='left' , on=['customer_number', 'new_yearMonth'])
```

```python
df_merged
```

```python
df_merged['Backlog_Sales'] = df_merged['Backlog_Sales'].fillna(0)
```

```python
df_merged
```

```python
df_merged['Monthly_Sum_Sales'] = df_merged['Monthly_Sum_Sales'] + df_merged['Backlog_Sales']
```

```python
df_merged
```

```python
df_merged['Monthly_Gross_Margin'] = df_merged['Monthly_Gross_Margin'] + df_merged['Backlog_Sales']
```

```python
df_merged
```

```python
df_merged.to_csv('GB_Cleaned_dataset.csv', index=False)
```
